i want you to built a dataset of 2 vectors, i want the dataset to have 10 sampls of tempeture in celius and the corssponding result in fahrenheit

```python
import numpy as np

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Generating 10 random samples of temperatures in Celsius
celsius_temperatures = np.random.uniform(low=-20.0, high=40.0, size=(10,))

# Converting Celsius temperatures to Fahrenheit
fahrenheit_temperatures = celsius_to_fahrenheit(celsius_temperatures)

# Creating the dataset
temperature_dataset = list(zip(celsius_temperatures, fahrenheit_temperatures))

# Displaying the dataset
for celsius, fahrenheit in temperature_dataset:
    print(f"Celsius: {celsius:.2f}°C  |  Fahrenheit: {fahrenheit:.2f}°F")
```

now i want you to create keras leyar with only one neuron that gets as an input the 2 vectors we created. in adittion i want to use mse for loss and learninh rate of 0.01 , also i want 500 epochs with batch of 1 and use adam opzimsation . in the end i want you to create a prediction on numbers not in the dataset.

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Generating the dataset
def generate_dataset():
    np.random.seed(42)
    celsius_temperatures = np.random.uniform(low=-20.0, high=40.0, size=(10,))
    fahrenheit_temperatures = (celsius_temperatures * 9/5) + 32
    return celsius_temperatures, fahrenheit_temperatures

celsius_temperatures, fahrenheit_temperatures = generate_dataset()

# Creating the Keras model
model = Sequential()
model.add(Dense(units=1, input_shape=(2,)))

# Compiling the model
opt = Adam(learning_rate=0.01)
model.compile(optimizer=opt, loss='mean_squared_error')

# Training the model
model.fit(x=np.column_stack((celsius_temperatures, fahrenheit_temperatures)),
          y=fahrenheit_temperatures,
          epochs=500,
          batch_size=1)

# Making predictions on numbers not in the dataset
new_celsius_temperatures = np.array([15.0, 25.0, 5.0])
predictions = model.predict(np.column_stack((new_celsius_temperatures, np.zeros_like(new_celsius_temperatures))))

# Displaying the predictions
for celsius, prediction in zip(new_celsius_temperatures, predictions):
    print(f"Celsius: {celsius:.2f}°C  |  Predicted Fahrenheit: {prediction[0]:.2f}°F")

```

better model with a much larger dataset and diffrent hyperparameters.

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split

# Generating the original dataset
def generate_dataset(size):
    np.random.seed(42)
    celsius_temperatures = np.random.uniform(low=-20.0, high=40.0, size=(size,))
    fahrenheit_temperatures = (celsius_temperatures * 9/5) + 32
    return celsius_temperatures, fahrenheit_temperatures

# Generating a larger dataset
dataset_size = 100000
celsius_temperatures, fahrenheit_temperatures = generate_dataset(dataset_size)

# Splitting the dataset into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(celsius_temperatures, fahrenheit_temperatures, test_size=0.2, random_state=42)

# Creating the Keras model with regularization
model = Sequential()
model.add(Dense(units=16, input_shape=(1,), activation='relu'))
model.add(Dropout(0.2))  # Adding dropout for regularization
model.add(Dense(units=1))

# Compiling the model
opt = Adam(learning_rate=0.001)
model.compile(optimizer=opt, loss='mean_squared_error')

# Training the model
history = model.fit(x=X_train, y=y_train, validation_data=(X_val, y_val),
                    epochs=200, batch_size=64, verbose=0)

# Making predictions on numbers not in the dataset
new_celsius_temperatures = np.array([15.0, 25.0, 5.0])
predictions = model.predict(new_celsius_temperatures)

# Displaying the differences
for celsius, prediction in zip(new_celsius_temperatures, predictions):
    actual_fahrenheit = (celsius * 9/5) + 32
    difference = prediction[0] - actual_fahrenheit
    percent_difference = (difference / actual_fahrenheit) * 100
    print(f"Celsius: {celsius:.2f}°C  |  Predicted Fahrenheit: {prediction[0]:.2f}°F  |  Actual Fahrenheit: {actual_fahrenheit:.2f}°F  |  Difference: {difference:.2f}°F  |  Percent Difference: {percent_difference:.2f}%")
```
